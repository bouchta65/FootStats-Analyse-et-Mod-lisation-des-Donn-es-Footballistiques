```python
from sqlalchemy import create_engine, text
engine = create_engine("postgresql://postgres:123@localhost:5432/Scaping_Data")

with engine.begin() as conn:
    conn.execute(text("""
                      CREATE TABLE IF NOT EXISTS Competition(
                      id_competition SERIAL PRIMARY KEY, 
                      nomcompetition VARCHAR(30)
                      )
                      """))
    conn.execute(text("""
                      CREATE TABLE IF NOT EXISTS Saison(
                      id_saison SERIAL PRIMARY KEY,
                      annee VARCHAR(30)
                      )
                      """))
    conn.execute(text("""
        CREATE TABLE IF NOT EXISTS Equipe(
            id_equipe int PRIMARY KEY,
            nomequipe VARCHAR(50),
            id_competition INT REFERENCES Competition(id_competition),
            id_saison INT REFERENCES Saison(id_saison)
        )
    """))

    conn.execute(text("""
        CREATE TABLE IF NOT EXISTS Joueur(
            id_joueur int PRIMARY KEY,
            nomjoueur VARCHAR(50),
            position VARCHAR(20),
            nationalite VARCHAR(30),
            id_equipe INT REFERENCES Equipe(id_equipe)
        )
    """))

    conn.execute(text("""
        CREATE TABLE IF NOT EXISTS Match(
            id_match int PRIMARY KEY,
            date_match DATE,
            heure TIME,
            round VARCHAR(20),
            venue VARCHAR(50),
            id_teamhome INT REFERENCES Equipe(id_equipe),
            id_team_away INT REFERENCES Equipe(id_equipe),
            id_competition INT REFERENCES Competition(id_competition),
            id_saison INT REFERENCES Saison(id_saison)
        )
    """))

    conn.execute(text("""
        CREATE TYPE resultat_enum AS ENUM ('Victoire', 'DÃ©faite', 'Nul');
        CREATE TABLE IF NOT EXISTS ResultatMatch(
            id_resultat SERIAL PRIMARY KEY,
            id_match INT REFERENCES Match(id_match),
            id_equipe INT REFERENCES Equipe(id_equipe),
            butsmarques INT,
            butsconcedes INT,
            resultat resultat_enum
        )
    """))

    conn.execute(text("""
        CREATE TABLE IF NOT EXISTS StatistiqueJoueur(
            id_stats SERIAL PRIMARY KEY,
            id_joueur INT REFERENCES Joueur(id_joueur),
            buts INT,
            passesdecisives INT,
            nbmatchesplayed INT,
            cartonsjaunes INT,
            cartonsrouges INT
        )
    """))
```
