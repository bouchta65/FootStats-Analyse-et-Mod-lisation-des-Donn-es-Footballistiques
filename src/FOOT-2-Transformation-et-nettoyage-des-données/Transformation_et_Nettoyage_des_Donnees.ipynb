### Importation des données des joueurs et des matchs avec Pandas

```python
import pandas as pd
players = "../../data/raw_data/players.csv"
matches = "../../data/raw_data/matches.csv"

players_df = pd.read_csv(players)
matches_df = pd.read_csv(matches)

players_df.head(30)
```

### Détection et suppression des doublons dans le DataFrame des joueurs et des matches

```python
players_df[players_df.duplicated()]
players_df.drop_duplicates(inplace=True)
```

```python
matches_df[matches_df.duplicated()]
matches_df.drop_duplicates(inplace=True)
```

### Suppression des lignes entièrement vides dans le DataFrame des joueurs et des matches

```python
players_df[players_df.drop(columns=['Team', 'Player']).isnull().all(axis=1)]
players_df = players_df[~players_df.drop(columns=['Team', 'Player']).isnull().all(axis=1)]
```

```python
matches_df[matches_df.drop(columns=['Team', 'Date']).isnull().all(axis=1)]
matches_df = matches_df[~matches_df.drop(columns=['Team', 'Date']).isnull().all(axis=1)]
```

```python
players_df.info()
```

### Inspection du DataFrame des  et des joueurs et nettoyage des noms de colonnes

```python
matches_df.columns = matches_df.columns.str.strip().str.replace(" ", "_")
matches_df.info()

```

### Séparation et nettoyage des champs de données

```python
matches_df['Time'] = matches_df['Time'].str.split(' ').str[0]
matches_df['GF'] = matches_df['GF'].str.split(' ').str[0]
matches_df['GA'] = matches_df['GA'].str.split(' ').str[0]
players_df['Nation'] = players_df['Nation'].str.split(' ').str[1]
```

```python
matches_df.info()
```

### Conversion des types de données pour l'analyse

```python
players_df["Age"] = players_df["Age"].astype(int)
players_df["MP"] = players_df["MP"].astype(int)
players_df["Starts"] = players_df["Starts"].astype(int)
players_df["Gls"] = players_df["Gls"].astype(int)
players_df["Ast"] = players_df["Ast"].astype(int)
players_df["G+A"] = players_df["G+A"].astype(int)
players_df["G-PK"] = players_df["G-PK"].astype(int)
players_df["PK"] = players_df["PK"].astype(int)
players_df["PKatt"] = players_df["PKatt"].astype(int)
players_df["CrdY"] = players_df["CrdY"].astype(int)

matches_df["Poss"] = matches_df["Poss"].astype(int)
matches_df["Date"] = pd.to_datetime(matches_df["Date"])
```

```python
players_df.info()
```

### Remplacement des valeurs manquantes par zéro

```python
players_df = players_df.fillna(0)
matches_df = matches_df.fillna(0)
```

### Assignation d’ID uniques aux équipes

```python
team_ids = {} 
id_counter = 1
for team in players_df["Team"].unique():
    team_ids[team] = id_counter
    id_counter += 1
    
players_df["id_equipe"] = players_df["Team"].map(team_ids)
```

### Assignation d’ID uniques aux compétitions

```python
comp_ids = {} 
id_counter = 1
for team in matches_df["Comp"].unique():
    comp_ids[team] = id_counter
    id_counter += 1
    
matches_df["id_competition"] = matches_df["Comp"].map(comp_ids)
```

### Assignation d’identifiants uniques aux joueurs et aux matchs

```python
players_df['id_joueur'] = range(1, len(players_df)+1)
matches_df['id_matche'] = range(1, len(matches_df)+1)
```

### Exportation des DataFrames nettoyés vers des fichiers CSV

```python
players_df.to_csv("../../data/processed_data/players.csv",index=False)
matches_df.to_csv("../../data/processed_data/matches.csv",index=False)
```
